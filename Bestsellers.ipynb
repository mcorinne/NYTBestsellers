```python
def CreateDataFrame(url):
    r = requests.get(url)
    r.json()
    pf = pd.DataFrame(r.json())
    books = pf.iloc[1]
    x = pd.DataFrame(books['results'])
    return(x)
```

```python
current_paperback = CreateDataFrame(url1)
current_hardcover = CreateDataFrame(url2)
oct_2021_pap = CreateDataFrame(url3)
oct_2020_pap = CreateDataFrame(url4)
oct_2019_pap = CreateDataFrame(url5)
oct_2018_pap = CreateDataFrame(url6)
oct_2017_pap = CreateDataFrame(url7)
oct_2016_pap = CreateDataFrame(url8)
oct_2015_pap = CreateDataFrame(url9)
```
